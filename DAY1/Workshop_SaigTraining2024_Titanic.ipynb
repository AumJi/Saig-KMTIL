# Fillna

```python
avg_age = train_data['Age'].mean()
train_data['Age'].fillna(avg_age, inplace=True)
print(f"Filling NaN with average age: {avg_age:.2f}")
```

# Add Column (Category)

```python
child_age = 18
train_data['AgeCategory'] = train_data['Age'].apply(lambda x : 'Child' if x <= child_age else 'Adult') # lambda คือ การเขียน function ฉบับย่อ โดยไม่ต้องมีชื่อฟังช์กัน
train_data
```
