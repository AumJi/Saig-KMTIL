# Clean Data

```python
train_data = pd.read_csv('titanic/train.csv')
train_data.head()
```

# Dropna

```python
train_data_dropna = train_data
train_data_dropna.info()
```

# Fillna

```python
avg_age = train_data['Age'].mean()
train_data['Age'].fillna(avg_age, inplace=True)
print(f"Filling NaN with average age: {avg_age:.2f}")
```

# Add Column (Category)

```python
child_age = 18
train_data['AgeCategory'] = train_data['Age'].apply(lambda x : 'Child' if x <= child_age else 'Adult') # lambda คือ การเขียน function ฉบับย่อ โดยไม่ต้องมีชื่อฟังช์กัน
train_data
```

# Train Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch"]
df = pd.get_dummies(train_data[features])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df,train_data['Survived'],test_size=0.25,random_state=42)
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)
acc=DT.score(x_test, y_test)
print(f"Acc: {acc:.4f}")
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train, y_train)
acc=RF.score(x_test, y_test)
print(f"Acc: {acc:.4f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(x_train,y_train)
acc = KNN.score(x_test,y_test)
print(f"Acc: {acc:.4f}")
```

```python
from sklearn.svm import SVC
SVC = SVC()
SVC.fit(x_train,y_train)
acc = SVC.score(x_test,y_test)
print(f"Acc: {acc:.4f}")
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
acc = LR.score(x_test,y_test)
print(f'Acc: {acc:.4f}')
```

```python
%%time
y_pred = RF.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
# เอ๋อ
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, linewidths=1,)
plt.ylabel("Actual")
plt.xlabel("Predict")
plt.show()
```
